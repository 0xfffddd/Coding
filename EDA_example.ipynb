```python
# EDA for example on Task list 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 0) Load data
# -----------------------------
path = "supply_chain_data.csv"
df = pd.read_csv(path)

print("✅ Loaded:", path)
print("Shape:", df.shape)
display(df.head())

# -----------------------------
# 1) Basic info
# -----------------------------
print("\n=== Columns ===")
print(df.columns.tolist())

print("\n=== Dtypes ===")
print(df.dtypes)

# -----------------------------
# 2) Data quality checks
# -----------------------------
print("\n=== Missing values ===")
missing = df.isnull().sum().sort_values(ascending=False)
display(missing[missing > 0])

print("\nTotal missing cells:", int(df.isnull().sum().sum()))

print("\n=== Duplicate rows ===")
dup_count = df.duplicated().sum()
print("Duplicate rows:", int(dup_count))

# -----------------------------
# 3) Separate numeric vs categorical
# -----------------------------
num_cols = df.select_dtypes(include=np.number).columns.tolist()
cat_cols = df.select_dtypes(exclude=np.number).columns.tolist()

print("\nNumeric columns:", num_cols)
print("Categorical columns:", cat_cols)

# -----------------------------
# 4) Summary stats
# -----------------------------
print("\n=== Numeric summary (describe) ===")
display(df[num_cols].describe().T)

print("\n=== Categorical summary (top counts) ===")
for c in cat_cols:
    print(f"\n--- {c} ---")
    display(df[c].value_counts(dropna=False).head(10))

# -----------------------------
# 5) Univariate plots
# -----------------------------
# 5.1 Price distribution
if "Price" in df.columns:
    plt.figure(figsize=(8,4))
    plt.hist(df["Price"].dropna(), bins=20)
    plt.title("Distribution of Price")
    plt.xlabel("Price")
    plt.ylabel("Frequency")
    plt.show()

# 5.2 Revenue distribution
if "Revenue generated" in df.columns:
    plt.figure(figsize=(8,4))
    plt.hist(df["Revenue generated"].dropna(), bins=20)
    plt.title("Distribution of Revenue generated")
    plt.xlabel("Revenue generated")
    plt.ylabel("Frequency")
    plt.show()

# 5.3 Defect rates distribution
if "Defect rates" in df.columns:
    plt.figure(figsize=(8,4))
    plt.hist(df["Defect rates"].dropna(), bins=20)
    plt.title("Distribution of Defect rates")
    plt.xlabel("Defect rates")
    plt.ylabel("Frequency")
    plt.show()

# 5.4 Example: Product type frequency (bar)
if "Product type" in df.columns:
    counts = df["Product type"].value_counts().head(10)
    plt.figure(figsize=(8,4))
    plt.bar(counts.index.astype(str), counts.values)
    plt.title("Top Product types (count)")
    plt.xticks(rotation=30, ha="right")
    plt.ylabel("Count")
    plt.show()

# -----------------------------
# 6) Correlation analysis (numeric only)
# -----------------------------
if len(num_cols) >= 2:
    corr = df[num_cols].corr(numeric_only=True)

    plt.figure(figsize=(10,8))
    plt.imshow(corr)
    plt.xticks(range(len(num_cols)), num_cols, rotation=90)
    plt.yticks(range(len(num_cols)), num_cols)
    plt.title("Correlation Heatmap (Numeric Features)")
    plt.colorbar()
    plt.tight_layout()
    plt.show()

    # Top absolute correlations (excluding self-correlation)
    corr_abs = corr.abs()
    np.fill_diagonal(corr_abs.values, 0)

    # flatten and sort
    pairs = (
        corr_abs.stack()
        .sort_values(ascending=False)
        .reset_index()
        .rename(columns={"level_0":"Var1","level_1":"Var2",0:"AbsCorr"})
    )

    print("\n=== Top 10 strongest correlations (absolute) ===")
    display(pairs.head(10))

else:
    print("Not enough numeric columns for correlation analysis.")

# -----------------------------
# 7) Quick PPT-ready insights (simple prints)
# -----------------------------
print("\n=== Quick Insights (PPT-friendly) ===")
print(f"- Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print(f"- Missing cells: {int(df.isnull().sum().sum())}")
print(f"- Duplicate rows: {int(dup_count)}")
if "Price" in df.columns:
    print(f"- Price: mean={df['Price'].mean():.2f}, min={df['Price'].min():.2f}, max={df['Price'].max():.2f}")
if "Revenue generated" in df.columns:
    print(f"- Revenue generated: mean={df['Revenue generated'].mean():.2f}, min={df['Revenue generated'].min():.2f}, max={df['Revenue generated'].max():.2f}")
if "Lead time" in df.columns:
    print(f"- Lead time: mean={df['Lead time'].mean():.2f}, min={df['Lead time'].min():.2f}, max={df['Lead time'].max():.2f}")

print("\n✅ Done.")
```
